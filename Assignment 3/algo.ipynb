```python
import numpy as np
import pandas as pd
```

```python
# Imporintg the dataset
data = pd.read_csv('Mall_Customers.csv')
data.head()
```

```python
clustering_data = data.iloc[:, [3, 4]].values
```

```python
clustering_data.shape
```

```python
import matplotlib.pyplot as plt
plt.scatter(clustering_data[:, 0], clustering_data[:, 1])
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
```

```python
# dbscan
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=3, min_samples=4)
model = dbscan.fit(clustering_data)
labels = model.labels_
```

```python
# unique labels
np.unique(labels)
```

```python
from sklearn import metrics
score = metrics.silhouette_score(clustering_data, labels)
score
```

```python
# plot the cluster with the color
plt.scatter(clustering_data[:, 0], clustering_data[:, 1], c=labels)
```
